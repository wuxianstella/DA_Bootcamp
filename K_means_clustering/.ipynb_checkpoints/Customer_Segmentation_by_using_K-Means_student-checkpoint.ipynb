#### Customers segmentation with K-means

[K-Means Sklean Link](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

Based on the scatter plot, we found that Spending score and Annual income might be useful to segment customers.

Looking at the scatter plot of **Spending score and Annual income**, it seems like 5 groups of customes. However, we should prove it. There are two most used methods to determine optimal clusters amount - **Elbow method and Silhouette method**.

##### Normalization of data
Before clustering, the data should be normalized (i.e., scaled) to make variables comparable. If we have a set of variables with incomparable units, the results of clustering may be biased. The clusters formed may be dependent on one variable significantly more than the other.

```python
from sklearn.preprocessing import StandardScaler
# define standard scaler
scaler = StandardScaler()
# transform data
df_scaled = scaler.fit_transform(df[["Spending Score (1-100)", "Annual Income (k$)"]])
```

```python
df_scaled
```

- **Elbow Method**

Elbow Method is an empirical method to find the optimal number of clusters for a dataset. In this method, we pick a range of candidate values of k, then apply K-Means clustering using each of the values of k. Find the average distance of each point in a cluster to its centroid, and represent it in a plot. Pick the value of k, where the average distance falls suddenly.

*If the line chart resembles an arm, then the “elbow” (the point of inflection on the curve) is a good indication that the underlying model fits best at that point*

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_scaled)
    distortions.append(kmeanModel.inertia_)
    
plt.figure(figsize=(10,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()

#inertia_: Sum of squared distances of samples to their closest cluster center.
```

##### Elbow method shows 5 is optimal.


- **Silhouette Method**

The silhouette Method is also a method to find the optimal number of clusters and interpretation and validation of consistency within clusters of data. The silhouette method computes silhouette coefficients of each point that measure how much a point is similar to its own cluster compared to other clusters. by providing a succinct graphical representation of how well each object has been classified. 

It's values are within [-1;1]. Optimal value is a peak.

Steps to find the silhouette coefficient of an i’th point:

- Compute a(i): The average distance of that point with all other points in the same clusters.

- Compute b(i): The average distance of that point with all the points in the closest cluster to its cluster.

- Compute s(i) — silhouette coefficient or i’th point using below mentioned formula.



- After computing the silhouette coefficient for each point, average it out to get the silhouette score.

```python
s_scores = []
clusters = range(2,10)
clusters_inertia = []

for n in clusters:
    KM_est = KMeans(n_clusters=n, init='k-means++').fit(df_scaled)
    clusters_inertia.append(KM_est.inertia_)   
    silhouette_avg = silhouette_score(df_scaled, KM_est.labels_)
    s_scores.append(silhouette_avg)
    
    
fig, ax = plt.subplots(figsize=(12,5))
ax = sns.lineplot(clusters, s_scores, marker='o', ax=ax)

ax.set_xlabel("Number of clusters")
ax.set_ylabel("Silhouette score")
plt.title('The Silhouette Method showing the optimal k')
plt.grid()
plt.show()
```

##### Silhouette Method shows 5 is optimal.

Elbow Method and the Silhouette Method are not like alternatives to each other for finding the optimal amount of clusters. Rather they are instruments for using together for a more confident decision.

```python
kmeanModel = KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans= kmeanModel.fit_predict(df_scaled)
plt.figure(figsize=(8,8))
plt.scatter(df_scaled[y_kmeans == 0, 0], df_scaled[y_kmeans == 0, 1], s = 100, c = 'g', label = 'Cluster 1')
plt.scatter(df_scaled[y_kmeans == 1, 0], df_scaled[y_kmeans == 1, 1], s = 100, c = 'b', label = 'Cluster 2')
plt.scatter(df_scaled[y_kmeans == 2, 0], df_scaled[y_kmeans == 2, 1], s = 100, c = 'r', label = 'Cluster 3')
plt.scatter(df_scaled[y_kmeans == 3, 0], df_scaled[y_kmeans == 3, 1], s = 100, c = 'burlywood', label = 'Cluster 4')
plt.scatter(df_scaled[y_kmeans == 4, 0], df_scaled[y_kmeans == 4, 1], s = 100, c = 'green', label = 'Cluster 5')
plt.scatter(kmeanModel.cluster_centers_[:, 0], kmeanModel.cluster_centers_[:, 1], s = 200, c = 'black', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

Groups could be described as:

- Cluster 1: Light green - low annual income and high spending score
- Cluster 4: Beige - high income and high spending score
- Cluster 3: Red - medium income and medium spending score
- Cluster 2: Blue - high income and low spending rate 
- Cluster 5: Green - low income and low spending score
